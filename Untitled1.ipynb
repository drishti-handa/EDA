```python
df['family_size'] = df['SibSp'] + df['Parch'] + 1
df
```

```python
def family_size_descp(a):
  if a == 1:
    return 'alone'
  elif (a > 1) and (a <= 5):
    return 'small'
  else:
    return 'large'
```

```python
df['family_type'] = df['family_size'].apply(family_size_descp)
df
```

```python
pd.crosstab(df['Survived'],df['family_type'],normalize='columns')*100
```
